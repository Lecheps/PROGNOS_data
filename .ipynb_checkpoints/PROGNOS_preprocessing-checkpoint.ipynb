## Saving meteo data

* u10: east-west wind speed m/s
* v10: north south wind speed m/s
* mslp: mean sea level pressure bar -> Not available, using 1013 as a placeholder
* t2: temperature degree Celsius
* hum: Relative humidity %
* cc: cloud cover -> Not available, using 0.6 as a placeholder

The results will be saved in *Langtjern\_weather.dat*


```python
#Saving the meteo data
GOTM_weather = allData['weather'].iloc[:,[4,4,6,1,3,0]]
GOTM_weather.index.name = 'date'
GOTM_weather.columns = ['u10', 'v10', 'mslp', 't2', 'hum', 'cc'] #Temporarily storing wind direction in mslp
GOTM_weather.loc[:,'u10'] = GOTM_weather['u10'] * np.cos(np.deg2rad(GOTM_weather['mslp']))
GOTM_weather.loc[:,'v10'] = GOTM_weather['v10'] * np.sin(np.deg2rad(GOTM_weather['mslp']))
GOTM_weather.loc[:,'mslp'] = 1013.0   
GOTM_weather.loc[:,'cc'] = 0.6 
GOTM_weather.loc[:,:].interpolate(inplace=True) #Linear interpolation of missing values and NaN's

#Plotting variables
fig, axes = plt.subplots(nrows = 4, ncols = 1, figsize = (20, 10))
GOTM_weather.loc[:,['u10','v10']].plot(ax = axes[0])
GOTM_weather.loc[:,'mslp'].plot(ax = axes[1],legend=True)
GOTM_weather.loc[:,'hum'].plot(ax = axes[2],legend=True)
axes[2].set_ylim(0,110)
GOTM_weather.loc[:,'cc'].plot(ax = axes[3],legend=True)


#Setting up formatting for different for saving as a .dat file
GOTM_weather.loc[:,'mslp'] = GOTM_weather['mslp'].map(lambda x : '%0.1f' % x)
GOTM_weather.loc[:,'hum'] = GOTM_weather['hum'].map(lambda x : '%0.1f' % x)
GOTM_weather.loc[:,'cc'] = GOTM_weather['cc'].map(lambda x : '%0.1f' % x)

#Saving    
GOTM_weather.to_csv('langtjern-weather.dat',sep = "\t",float_format='%0.2f',dateformat = '%Y-%m-%d %H:%M:%S')
```

## Saving inlet data

Discharge is obtained from stage with the use of the following rating curve:

 $ \textrm{if } gh > 0.345 \textrm{ then }  q = 2.391 \times (gh - 0.345)^{2.5} \textrm{ else } q = 0.0001 $
 
 $q(m^3/s) \textrm { and } gh(m) $
 
**Please note** In the info received from NVE the value for the multiplier was 2391, which yielded absurd discharge. This was changed to 2.391
 
**Please note** The inlet just covers a portion of the catchment and will be scaled by area in order to be representative of the entire catchment. This is a kludge.

CDOM data prior to 2012 are highly uncertain. It is recommended not to use them.
 
 Two data files will be saved:
 
 * CDOM data in $\mu g/l$
 * Discharge and water temperature in $m^3/s$ and $^\circ C$ respectively 
 
 
 

```python
#Filling in missing values and nans from inlet data using linear interpolation
allData['inlet'].interpolate(inplace=True)

#Transforming stage into discharge
inletQ = lambda x : 2.391* np.power(x - 0.345, 2.5) if (x > 0.345) else 0.0001
allData['inlet'] = allData['inlet'].rename(columns={'Vannstand_m':'Discharge',u'CDOM_µg/L' : 'CDOM'})
allData['inlet'].loc[:,'Discharge'] = allData['inlet'].loc[:,'Discharge'].apply(inletQ)
allData['inlet'].loc[:,'Discharge'] = allData['inlet'].loc[:,'Discharge'] * (4.8 + 0.23) / 0.8

print("Th average discharge is: ")
print (np.mean(allData['inlet'].loc[:,'Discharge']))

#Visualizing data
fig, axes = plt.subplots(nrows = 3, ncols = 1, figsize = (20, 15))
allData['inlet'].loc[:,u'CDOM'].plot(ax = axes[0],legend=True)
allData['inlet'].loc[:,'Temperatur_C'].plot(ax = axes[1],legend=True)
allData['inlet'].loc[:,'Discharge'].plot(ax = axes[2],legend=True)

#Saving inlet data
fcdom = open('langtjern-inlet-cdom-ug-per-l.dat','w')
fdisch = open('langtjern-inlet-q.dat','w')

for i,j in allData['inlet'].iterrows() :
    fcdom.write(datetime.datetime.strftime(i,'%Y-%m-%d %H:%M:%S') + ' ' + str(j.loc['CDOM']) + '\n')    
    fdisch.write(datetime.datetime.strftime(i,'%Y-%m-%d %H:%M:%S') + ' ' + str(j.loc['Discharge']) + ' ' + str(j.loc['Temperatur_C']) + '\n') 

fcdom.close()
fdisch.close()
```

## Saving outlet data

Discharge is obtained from the stage using the following rating curve:

 $ \textrm{if } gh > 0.37 \textrm{ then }  q = 3.2136 \times gh^{2.453} \textrm{ else } q = 3.794 \times gh^{2.692}$
 
$q(m^3/s) \textrm { and } gh(m) $

**Please note** The gauge height is not directly measured at the flume but at some upstream point and the gauge height should be corrected by a height (substraction) of *0.1403*

CDOM at the outlet must be corrected because of recalibration in 2012 $10\mu g/l$ should be added to all values before _22/11/2012_ NO time.

Additionaly with respect to the inlet, the conductivity is measured here as well.

Three data files will be saved:
* CDOM data in  μg/lμg/l 
* Discharge and water temperature in  m3/s  and  ∘C respectively
* Conductivity in mS/m

```python
#Filling in missing values and nans from inlet data using linear interpolation
allData['outlet'].interpolate(inplace=True) 

#Transforming stage into discharge
outletQ = lambda x : 3.2136 * np.power(x - 0.1403, 2.453) if x > 0.37 else 3.794 * np.power(x - 0.1403, 2.692)  

allData['outlet'] = allData['outlet'].rename(columns={'Vannstand_m':'Discharge',u'CDOM_µg/L' : 'CDOM'})
allData['outlet'].loc[:,'Discharge'].apply(outletQ)

print("The average discharge is: ")
print (np.mean(allData['outlet'].loc[:,'Discharge']))

#Visualizing data
fig, axes = plt.subplots(nrows = 5, ncols = 1, figsize = (20, 10))
allData['outlet'].loc[:,u'CDOM'].plot(ax = axes[0],legend=True)
allData['outlet'].loc[:,'Konduktivitet_mS/m'].plot(ax = axes[1],legend=True)
allData['outlet'].loc[:,'pH'].plot(ax = axes[2],legend=True)
allData['outlet'].loc[:,'Temperatur_C'].plot(ax = axes[3],legend=True)
allData['outlet'].loc[:,'Discharge'].plot(ax = axes[4],legend=True)

#Saving outlet data
fcdom = open('langtjern-outlet-cdom-ug-per-l.dat','w')
fdisch = open('langtjern-outlet-q.dat','w')
fcond = open('langtjern-outlet-cond-us-per-m.dat','w')
fpH = open('langtjern-outlet_ph.dat','w')

for i,j in allData['outlet'].iterrows() :
    fcdom.write(datetime.datetime.strftime(i,'%Y-%m-%d %H:%M:%S') + ' ' + str(j.loc[u'CDOM']) + '\n')    
    fdisch.write(datetime.datetime.strftime(i,'%Y-%m-%d %H:%M:%S') + ' -' + str(j.loc['Discharge']) + ' ' + str(j.loc['Temperatur_C']) + '\n')
    fcond.write(datetime.datetime.strftime(i,'%Y-%m-%d %H:%M:%S') + ' ' + str(j.loc['Konduktivitet_mS/m']) + '\n') 
    fpH.write(datetime.datetime.strftime(i,'%Y-%m-%d %H:%M:%S') + ' ' + str(j.loc['pH']) + '\n')
fcdom.close()
fdisch.close()
```

## Saving profile data

GOTM will fill in the missing values in the profile data. Therefore only the depths at which there actually are measurements are included here.

Two profiles will be saved:

* One for temperature in $^{\circ}C$
* One for $O_2$ in _%_

```python
#Storing profile data in an array for visualization
t = []
O2 = []
for i in allData['lake'].index.unique() :
    subset = allData['lake'].iloc[allData['lake'].index == i, :]
    subset = subset.values
    t.append(subset[:,3])
    if np.sum(subset[:,2]== subset[:,2]) == 2 :
        O2.append(subset[subset[:,2] == subset[:,2],2])   
        
O2 = np.matrix(O2)
O2 = O2.transpose()
#Visualization
plt.imshow(O2,interpolation='none', origin='lower',cmap=plt.cm.terrain,extent=(-10,30,-10,30))   

#Saving temperature and oxygen profiles
fid = open('langtjern-lake-temp.dat','w')
fox = open('langtjern-lake-O2-percentage.dat','w')

#Saving the temperature and oxygen profiles. Saving only depth and timesteps for which there are data
for i in allData['lake'].index.unique() :
    subset = allData['lake'].iloc[allData['lake'].index == i, :]#There has to be a more efficient way of doing this
    idx_O2 = subset.iloc[:,2] == subset.iloc[:,2]
    idx_temp = subset.iloc[:,3] == subset.iloc[:,3]
    num_O2 = np.int_(np.sum(idx_O2))
    num_temp = np.int_(np.sum(idx_temp))
    if num_O2 > 0 :
        fox.write(datetime.datetime.strftime(i,'%Y-%m-%d %H:%M:%S') + ' ' + str(num_O2) + ' ' + str(2) + '\n')
        subset_O2 = subset[idx_O2.values]
        for j,k in subset_O2.iterrows() : 
            fox.write('-' + str(k[1]) + ' ' +  str(k[2]) + '\n' )
    if num_temp > 0 :
        fid.write(datetime.datetime.strftime(i,'%Y-%m-%d %H:%M:%S') + ' ' + str(num_temp) + ' ' + str(2) + '\n')
        subset_temp = subset.iloc[idx_temp.values]
        for j,k in subset_temp.iterrows() :
            fid.write('-' + str(k[0]) + ' ' + str(k[3]) + '\n')
    
fid.close()
fox.close()
```
